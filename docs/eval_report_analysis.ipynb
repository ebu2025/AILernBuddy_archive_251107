# Evaluation Report Quickstart
This notebook demonstrates how to query the `/eval/report` endpoint and explore
normalised gains by topic and instructional strategy. It expects the FastAPI
application to be running locally (for example via `uvicorn app:app --reload`).

```python
import os
import requests

BASE_URL = os.environ.get('EVAL_API_BASE_URL', 'http://localhost:8000')
FILTERS = {
    'learner_id': None,  # e.g. 'learner-42'
    'topic': None,       # e.g. 'bpmn'
    'strategy': None,    # e.g. 'metacognitive'
}

response = requests.get(f"{BASE_URL}/eval/report", params={k: v for k, v in FILTERS.items() if v})
response.raise_for_status()
report = response.json()
report.keys()
```

```python
import pandas as pd

summary_df = pd.DataFrame(report['summary'])
summary_df.sort_values(['topic', 'strategy'], inplace=True)
summary_df
```

```python
pairs_df = pd.DataFrame(report['pairs'])
pairs_df[['learner_id', 'topic', 'pre_score', 'post_score', 'normalized_gain', 'strategy']]
```
